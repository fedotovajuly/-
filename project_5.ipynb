### Вводная часть

Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Для планирования рекламных компаний интернет - магазину требуется выявить определяющие успешность игры закономерности. Для этого в нашем распоряжении есть исторические данные до 2016 года из открытых источников о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Данные за 2016 год могут быть неполными.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

**Описание данных**

- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера. Первоначально "K-A" ("Kids to Adults"), соотвественно обозначение "K-A" следует заменить на "Е",

- "AO" ("Adults Only 18+") — "Только для взрослых": Содержание игры только для взрослых старше 18 лет. Продукты из данной категории могут содержать длительные сцены жестокого насилия и/или очень откровенное сексуальное содержимое.
- "EC" ("Early childhood") — "Для детей младшего возраста": Игра подходит для детей от 3 лет и не содержит материалов, которые родители могли бы счесть неподходящими. Продукты, получившие данный рейтинг, изначально разрабатываются для детей и обычно представляют собой развивающие игры.
- "RP" ("Rating Pending") — "Рейтинг ожидается": Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу.

2. В столбцах с оценками экспертов и пользователей и возрастным рейтингом большое количество пропусков. Увалив их, мы бы потерями большую часть данных, необходимых для анализа.При этом не имеет смысла их заполнять ни медианой не средним, так как в случае с оценками пользователей и критиков это личное мнение конкретных людей о играх. Его не усреднишь. 
3. Рейтинг от организации ESRB же так же решили не заполнять, так как здесь заполненные нами данные могут не отражать должным образом для какой аудитории предполагались игры с пропусков в данном столбце. Соответственно, **пропуски заменим "заглушкой".**

Пропуски в столбцах с названиями игр и годом релиза и жанром скорее всего обусловлены человеческим фактором. 

Пропуски в столбцах с оценками пользователей и критиков могут быть обусловлены отсутствием самих оценок, особенно это актуально для игр с годом релиза в периоде с 80-х по начало 2000-ных. А так же не стоит сбрасывать со счетов человеческий фактор и ошибки в записи данных, приводящие к пропускам. Еще немаловажно отметить, что пропуски в оценках закономерны для игр, которые еще не вышли.

Пропуски в столбце с возрастной категорией для игр с датой релиза до 2003 года могут быть обусловлены тем, что Entertainment Software Rating Board еще не существовала, соответственно возрастную категорию данная организация не могла присудить. К тому же, в датасете присутствуют игры, направленные на разные рынки сбыта, т.е. для разных стран, а ESRB оценивает игры, предназначенные для США и Канады. Но и про человеческий фактор как причину пропусков не забываем.

```python
for platform in ["ps4", "xone", "wiiu","x360","pc"]:
    actual[actual["platform"] == platform].plot(x='critic_score', y='number_of_sales', kind='scatter', alpha=0.6, grid = True)
    plt.title(f'Диаграмму рассеивания количества проданных копий к оценкам критиков для {platform}')
    plt.xlabel('оценка критиков')
    plt.ylabel('млн проданных копий')
    plt.show() 
    
    actual[actual["platform"] == platform].plot(x='user_score', y='number_of_sales', kind='scatter', alpha=0.6, grid = True)
    plt.title(f'Диаграмму рассеивания количества проданных копий к оценкам пользователей  для {platform}')
    plt.xlabel('оценка пользователей')
    plt.ylabel('млн проданных копий')
    plt.show() 
    
    print(platform, '- коэфициент корреляции количества продаж и оценкок критиков:', "{:.2f}".format(actual[actual['platform'] == platform]['number_of_sales'].corr(actual[actual['platform'] == platform]['critic_score'], method='spearman')))
    print(platform, '- коэфициент корреляции количества продаж и оценок пользователей:', "{:.2f}".format(actual[actual['platform'] == platform]['number_of_sales'].corr(actual[actual['platform'] == platform]['user_score'], method='spearman')))
```

### Проверим гипотезы

1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
2. Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Нулевая гипотеза (H0) - это утверждение, которое предполагает отсутствие различий или эффекта между выборками или группами и является базовой гипотезой для проверки. 

Альтернативная гипотеза (H1) - это утверждение, которое предполагает наличие различий или эффекта между выборками или группами и является противоположностью нулевой гипотезе.

Нулевая гипотеза для первой проверки звучит так: средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
Альтернативная гипотеза: средние пользовательские рейтинги платформ Xbox One и PC разные.

Для второй проверки:
Нулевая гипотеза: средние пользовательские рейтинги жанров Action и Sports одинаковые.
Альтернативная гипотеза: средние пользовательские рейтинги жанров Action и Sports разные.

В данном случае можно применить t-критерий Стьюдента для независимых выборок. Данный критерий используется для сравнения средних двух групп. Мы можем использовать этот критерий, так как у нас есть две независимые группы (пользовательские рейтинги на Xbox One и PC, а также рейтинги жанров Action и Sports), и мы хотим сравнить их средние значения.

В качестве порогового значения alpha выбираем 0.05. Это означает, что мы готовы считать результаты статистически значимыми, если вероятность случайной ошибки не превышает 5%.

Таким образом, мы применили t-критерий Стьюдента для проверки гипотез и выбрали пороговое значение alpha равным 0.05. Эти выборы основаны на стандартных подходах при проверке гипотез и помогут нам определить, являются ли наши гипотезы статистически значимыми.
Для проверки гипотез о равенстве двух генеральных совокупностей будем использовать метод scipy.stats.ttest_ind() 

#### Гипотеза о раверстве средних пользовательских рейтингов платформ Xbox One и PC.


Н0 (нулевая гипотеза): Средние пользовательские рейтинги платформ Xbox One и PC равны.

Н1 (альтернативная гипотеза): Средние пользовательские рейтинги платформ Xbox One и PC не равны.

Зададим уровень статитистической значимости = 0.05.

```python
xone_user_score = actual.query('platform == "xone"')
pc_user_score = actual.query('platform == "pc"')
xone1 = xone_user_score['user_score']
pc1 = pc_user_score['user_score']

# Уровень статистической значимости
alpha = 0.05

# Статистический t -  тест (nan_policy='omit': игнорирует nan в данных)
results = st.ttest_ind(xone1, pc1, nan_policy='omit')


print('p-значение', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Исходя из того, что не удалось отвергнуть нулевую гипотезу, нет оснований утверждать, что средние пользовательские рейтинги платформ Xbox One и PC значимо различаются.

#### Гипотеза о неравенстве средниих пользовательских рейтингов жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования»).

Н0 (нулевая гипотеза): Средние пользовательские рейтинги платформ жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») равны.

Н1 (альтернативная гипотеза): Средние пользовательские рейтинги платформ жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») не равны.

Зададим уровень статитистической значимости = 0.05.

```python
action_user_score = actual.query('genre == "action"')
sports_user_score = actual.query('genre == "sports"')
action1 = action_user_score['user_score']
sports2 = sports_user_score['user_score']

# Уровень статистической значимости
alpha = 0.05

# Статистический t -  тест (nan_policy='omit': игнорирует nan в данных)
results1 = st.ttest_ind(action1, sports2, nan_policy='omit')


print('p-значение', results1.pvalue)

if results1.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Наши данные не противоречат гипотезе: средние пользовательские рейтинги жанров Action и Sports действительно различны, поскольку отвергли нулевую гипотезу о равенстве средниих пользовательских рейтингов жанров Action и Sports.
